#### PLOTTING WITH MATPLOTLIB AND SEABORN

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv("Car_sales (1).csv",parse_dates=['Latest_Launch'])
data.head()
```

```python
data.info()
```

```python
data.rename(columns={'__year_resale_value':'year_resale_value'},inplace=True)
data.head()
```

```python
data.sort_values(by='Latest_Launch',inplace=True)
```

```python
data.head()
```

```python
data.reset_index(drop=True,inplace=True)
```

```python
data.head(10)
```

```python
data["year"]=data.Latest_Launch.dt.year
data["month"]=data.Latest_Launch.dt.month
data["day"]=data.Latest_Launch.dt.day
```

```python
data.describe()
```

```python
data.isnull().mean()*100
```

```python
data.dropna(inplace=True)
data.drop_duplicates(inplace=True)
data.shape
```

### UNIVARIATE ANALYSIS


#### HISTOGRAM

```python
sns.histplot(data=data,x='Sales_in_thousands')
plt.show()
```

#### BOXPLOT

```python
sns.boxplot(data=data,x='Sales_in_thousands')
plt.show()
```

```python
sns.boxplot(data=data,y='Sales_in_thousands')
plt.show()
```

#### kde plot

```python
sns.kdeplot(data=data.Sales_in_thousands)
plt.show()
```

### subplots

```python
sns.histplot(data=data)
plt.show()
```

```python
plt.figure(figsize=(12,7))
plt.xticks(rotation=75)
sns.boxplot(data=data)
plt.show()
```

```python
plt.figure(figsize=(9,6))
sns.kdeplot(data=data)
plt.show()
```

### FOR LOOP IN SUBPLOTS

```python
for i in data.columns:
    if data[i].dtype!='object':
        sns.boxplot(y=data[i])
        plt.show()
```

```python
for i in data.columns:
    if data[i].dtype!='object':
        sns.histplot(x=data[i])
        plt.show()
```

```python
for i in data.columns:
    if data [i].dtype!='object':
        sns.kdeplot(x=data[i])
        plt.show()
```

### subplots in matplotlib

```python
plt.figure(figsize=(12,14))
plt.subplot(2,2,1)
sns.boxplot(data=data,y='Sales_in_thousands',color='r')

plt.subplot(2,2,2)
sns.boxplot(data=data,y='year_resale_value',color='y')

plt.subplot(2,2,3)
sns.boxplot(data=data,y='Price_in_thousands',color='g')

plt.subplot(2,2,4)
sns.boxplot(data=data,y='Horsepower',color='b')

plt.show()
```

### line chart

```python
plt.figure(figsize=(12,6))
plt.xlabel('year resale value')
plt.ylabel('price')
plt.title('figure')
sns.lineplot(data=data,x='year_resale_value',y='Price_in_thousands')
plt.show()
```

### count plots


```python
plt.figure(figsize=(12,5))
plt.xticks(rotation=75)
sns.countplot(data=data,x='Manufacturer')
plt.show()
```

### area chart

```python
plt.figure(figsize=(18,5))
plt.xticks(rotation=75)
plt.stackplot(data.Latest_Launch,data.Sales_in_thousands)
plt.show()
```

### Horizontal bar chart

```python
plt.figure(figsize=(6,12))
plt.xticks(rotation = 90)
sns.barplot(data=data, x='Sales_in_thousands', y='Manufacturer')
plt.show()
```

### Vertical Bar plot

```python
plt.figure(figsize=(12,4))
plt.xticks(rotation = 90)
sns.barplot(data=data, x='Manufacturer', y='Horsepower')
plt.show()
```

### Scatter plot

```python
sns.scatterplot(data=data, x='Horsepower', y='Sales_in_thousands')
plt.show()
```

```python
sns.scatterplot(data=data, x='Horsepower', y='Sales_in_thousands', hue='Vehicle_type')
plt.show()
```

### Pie chart

```python
data.Vehicle_type.value_counts().index
```

```python
palette_color = sns.color_palette('dark')
```

```python
plt.pie(data=data,\
x=data.Vehicle_type.value_counts(),\
explode = [0.1, 0], \
colors=palette_color,\
labels=['Passenger', 'Car'],\
autopct='%0.2f%%');
```

### Correlation matrix or Heatmap

```python
sns.heatmap(data.corr()[["Price_in_thousands"]].sort_values(by="Price_in_thousands"))
plt.figure(figsize=(50,18))
plt.show()
```
