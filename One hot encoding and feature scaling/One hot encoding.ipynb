# One hot encoding and feature scaling

```python
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## Loading the dataset

```python
data=pd.read_csv('Titanic.csv')
data.head()
```

```python
data.dropna(inplace=True)
data.drop_duplicates(inplace=True)
```

```python
data.info()
```

```python
data.drop(columns={'Name','Ticket','Cabin','PassengerId'},inplace=True)
```

## One hot encoding

```python
data['SEX']=pd.get_dummies(data=data.Sex,drop_first=True)
```

```python
data.head()
```

## Converting the whole dataset

```python
data_no_dummies=pd.get_dummies(data=data)
```

```python
data_no_dummies
```

```python
data_no_dummies.shape
```

```python
data_with_dummies=pd.get_dummies(data=data,drop_first=True)
data_with_dummies
```

```python
data_with_dummies.shape
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler,RobustScaler,LabelEncoder
```

## Label encoder

```python
le=LabelEncoder()
```

```python
data
```

```python
data['EMBARKED']=le.fit_transform(data.Embarked)
```

```python
data
```

## Data Scaling or Data Normalisation

### Minmax scaler

```python
minmax=MinMaxScaler()
```

```python
data[['Age']]
```

```python
data[['Age']].values
```

```python
age=data[['Age']].values
```

```python
age
```

```python
data['AGE']=minmax.fit_transform(age)
```

```python
data.head()
```

```python
data.AGE.min()
```

```python
data.AGE.max()
```

## Standard scaler

```python
sc=StandardScaler()
```

```python
data['AGE_SC']=sc.fit_transform(age)
```

```python
data
```

```python
data.AGE_SC.min()
```

```python
data.AGE_SC.max()
```

## Robust scaler

```python
rc=RobustScaler()
```
